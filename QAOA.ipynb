```python
# Import necessary libraries
import numpy as np
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
# The function execute(...) from qiskit has been removed in Qiskit 1.0.
# Instead you now use transpile(...) followed by backend.run(...)
from qiskit.circuit import Parameter
from qiskit_ibm_runtime import QiskitRuntimeService
import matplotlib.pyplot as plt
```

```python
# Let's create our first quantum circuit
qc = QuantumCircuit(2) # quantum circuit with 2 qubits
#Add a Hadamard gate to create superposition 
qc.h(0) #Qubit 0 is not in |0> + |1> state
#Add a CNOT gate to create entaglement
qc.cx(0, 1)
print("Our first quantum circuit : ") 
print(qc)
```
